```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset desde el archivo CSV
ruta_deuda = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/DeudaSeguridadSocial/DeudaSeguridadSocial.csv'

deuda_df = pd.read_csv(ruta_deuda, delimiter=';', decimal=',')
# Verificar los primeros registros del dataset para asegurarnos de que se haya cargado correctamente
print(deuda_df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

deuda_df['Deuda SS'] = deuda_df['Deuda SS'].str.replace('.', '').str.replace(',', '.').astype(float)

# Crear la gráfica
plt.figure(figsize=(10, 6))

# Graficar la deuda de la seguridad social
plt.plot(deuda_df['Año'], deuda_df['Deuda SS'], label='Deuda SS')


# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Deuda de la Seguridad Social')
plt.title('Deuda de la Seguridad Social a lo largo de los años')

plt.legend()
plt.show()
```
