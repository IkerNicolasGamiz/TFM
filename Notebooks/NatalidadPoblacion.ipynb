```python
import csv
import pandas as pd

# Función para cargar los datos de la tasa de natalidad desde un archivo CSV
def cargar_datos_natalidad_desde_csv(archivo_csv):
    datos_natalidad = {}
    with open(archivo_csv, newline='') as csvfile:
        lector_csv = csv.reader(csvfile, delimiter=';')  
        encabezados = next(lector_csv)
        for fila in lector_csv:
            if len(fila) >= 2:
                año = int(fila[0])
                valor = float(fila[1].replace(',', '.'))
                datos_natalidad[año] = valor
            else:
                print("Advertencia: La fila no tiene suficientes elementos:", fila)
    return datos_natalidad

# Función para cargar los datos de población de menores de 15 años desde un archivo Excel
def cargar_datos_poblacion_menores15_desde_excel(archivo_excel):
    df = pd.read_excel(archivo_excel, engine='openpyxl')
    menores_de_15 = df.iloc[2:3, 1:]
    suma_por_año_menores_de_15 = menores_de_15.sum(axis=0)
    resultados_por_año = {}
    for año in df.columns[1:]:
        # Convertir año a entero eliminando las comillas
        año_int = int(año.replace("'", ""))
        resultados_por_año[año_int] = suma_por_año_menores_de_15[año]
    return resultados_por_año

# Rutas de los archivos de datos
archivo_csv_natalidad = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/TasaDeNatalidad/TasaDeNatalidad(1000 habitantes).csv'
archivo_excel_poblacion = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/PoblacionPorEdad/PoblacionPorEdad.xlsx'

# Cargar datos de la tasa de natalidad
datos_natalidad = cargar_datos_natalidad_desde_csv(archivo_csv_natalidad)

# Cargar datos de población de menores de 15 años
datos_poblacion_menores15 = cargar_datos_poblacion_menores15_desde_excel(archivo_excel_poblacion)

# Verificar los años comunes entre ambos conjuntos de datos
años_comunes = set(datos_natalidad.keys()) & set(datos_poblacion_menores15.keys())
print("Años comunes:", años_comunes)

# Combinar ambos conjuntos de datos en un solo diccionario
datos_combinados = {}
for año in años_comunes:
    datos_combinados[año] = {
        "Tasa de Natalidad": datos_natalidad[año] * 1000,
        "Población Menores de 15": datos_poblacion_menores15[año]
    }

# Mostrar los resultados combinados
for año, valores in datos_combinados.items():
    print(año)
    print(valores)
    print()
```

```python
from scipy.stats import pearsonr

# Obtener las listas de datos para el cálculo del coeficiente de correlación
tasa_de_natalidad = [datos_combinados[año]["Tasa de Natalidad"] for año in años_comunes]
poblacion_menores15 = [datos_combinados[año]["Población Menores de 15"] for año in años_comunes]

# Calcular el coeficiente de correlación de Pearson y su valor p
coeficiente_pearson, valor_p = pearsonr(tasa_de_natalidad, poblacion_menores15)

# Imprimir el resultado
print("Coeficiente de correlación de Pearson:", coeficiente_pearson)
print("Valor p:", valor_p)
```
