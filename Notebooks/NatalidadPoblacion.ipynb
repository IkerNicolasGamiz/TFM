```python
import csv
import pandas as pd

# Función para cargar los datos de la tasa de natalidad desde un archivo CSV
def cargar_datos_natalidad_desde_csv(archivo_csv):
    datos_natalidad = {}
    with open(archivo_csv, newline='') as csvfile:
        lector_csv = csv.reader(csvfile, delimiter=';')  
        encabezados = next(lector_csv)
        for fila in lector_csv:
            if len(fila) >= 2:
                año = int(fila[0])
                valor = float(fila[1].replace(',', '.'))
                datos_natalidad[año] = valor
            else:
                print("Advertencia: La fila no tiene suficientes elementos:", fila)
    return datos_natalidad

# Función para cargar los datos de la esperanza de vida desde un archivo CSV
def cargar_datos_esperanza_desde_csv(archivo_csv):
    datos_esperanza = {}
    with open(archivo_csv, newline='') as csvfile:
        lector_csv = csv.reader(csvfile, delimiter=';')  
        encabezados = next(lector_csv)
        for fila in lector_csv:
            if len(fila) >= 2:
                año = int(fila[0])
                valor = float(fila[1].replace(',', '.'))
                datos_esperanza[año] = valor
            else:
                print("Advertencia: La fila no tiene suficientes elementos:", fila)
    return datos_esperanza

# Función para cargar los datos de población de menores de 15 años desde un archivo Excel
def cargar_datos_poblacion_menores15_desde_excel(archivo_excel):
    df = pd.read_excel(archivo_excel, engine='openpyxl')
    menores_de_15 = df.iloc[2:5, 1:]
    suma_por_año_menores_de_15 = menores_de_15.sum(axis=0)
    resultados_por_año = {}
    for año in df.columns[1:]:
        # Convertir año a entero eliminando las comillas
        año_int = int(año.replace("'", ""))
        resultados_por_año[año_int] = suma_por_año_menores_de_15[año]
    return resultados_por_año

# Función para cargar los datos de población total desde un archivo Excel
def cargar_datos_poblacion_total_desde_excel(archivo_excel):
    df = pd.read_excel(archivo_excel, engine='openpyxl')
    poblacion_total = df.iloc[1:2, 1:]  # Seleccionar la fila correspondiente a la población total
    resultados_por_año = {}
    for año in df.columns[1:]:
        # Convertir año a entero eliminando las comillas
        año_int = int(año.replace("'", ""))
        resultados_por_año[año_int] = poblacion_total[año].iloc[0]
    return resultados_por_año
    
# Rutas de los archivos de datos
archivo_csv_natalidad = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/TasaDeNatalidad/TasaDeNatalidad(1000 habitantes).csv'
archivo_csv_esperanza = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/EsperanzaDeVida/EsperanzaDeVidaGeneral.csv'
archivo_excel_poblacion = 'C:/Users/Iker Nikolas/Desktop/TFM-24/Github/DatasetsNormalizados/PoblacionPorEdad/PoblacionPorEdad.xlsx'

# Cargar datos de la tasa de natalidad
datos_natalidad = cargar_datos_natalidad_desde_csv(archivo_csv_natalidad)

#Cargar datos de la esperanza de vida
datos_esperanza = cargar_datos_esperanza_desde_csv(archivo_csv_esperanza)

# Cargar datos de población de menores de 15 años
datos_poblacion_menores15 = cargar_datos_poblacion_menores15_desde_excel(archivo_excel_poblacion)

# Cargar datos de población total
datos_poblacion_total = cargar_datos_poblacion_total_desde_excel(archivo_excel_poblacion)

# Verificar los años comunes entre ambos conjuntos de datos
años_comunes = set(datos_natalidad.keys()) & set(datos_poblacion_menores15.keys())
print("Años comunes:", años_comunes)

datos_combinados = {}
for año in años_comunes:
    datos_combinados[año] = {
        "Tasa de Natalidad": datos_natalidad[año],
        "Esperanza de vida": datos_esperanza[año],
        "Población Menores de 15": datos_poblacion_menores15[año],
        "Población Total": datos_poblacion_total[año]
    }

# Mostrar los resultados combinados
for año, valores in datos_combinados.items():
    print(año)
    print(valores)
    print()
```

```python
#Tasa de Natalidad Normalizada= (Tasa de Natalidad × Poblacion total)/1000

import numpy as np
from scipy.stats import pearsonr

# Función para normalizar la tasa de natalidad para cada año
def normalizar_tasa_natalidad_por_año(datos_combinados):
    tasa_natalidad_normalizada_por_año = {}
    for año, datos in datos_combinados.items():
        tasa_natalidad_normalizada_por_año[año] = (datos['Tasa de Natalidad'] * datos['Población Total']) / 1000
    return tasa_natalidad_normalizada_por_año

# Normalizar la tasa de natalidad para cada año
tasa_natalidad_normalizada_por_año = normalizar_tasa_natalidad_por_año(datos_combinados)

# Actualizar el diccionario datos_combinados con la nueva tasa de natalidad normalizada
for año, datos in datos_combinados.items():
    datos['Tasa de Natalidad Normalizada'] = tasa_natalidad_normalizada_por_año[año]

# Imprimir los datos actualizados
for año, valores in datos_combinados.items():
    print(año)
    print(valores)
    print()
```

```python
from scipy.stats import pearsonr

# Obtener las listas de valores para las variables de interés
poblacion_total = [datos['Población Total'] for año, datos in datos_combinados.items()]
esperanza_vida = [datos['Esperanza de vida'] for año, datos in datos_combinados.items()]
poblacion_menores15 = [datos['Población Menores de 15'] for año, datos in datos_combinados.items()]
tasa_natalidad = [datos['Tasa de Natalidad Normalizada'] for año, datos in datos_combinados.items()]

# Calcular la correlación entre la población total y la esperanza de vida
correlacion_poblacion_esperanza, _ = pearsonr(poblacion_total, esperanza_vida)
print("Correlación entre población total y esperanza de vida:", correlacion_poblacion_esperanza)

# Calcular la correlación entre la población menor de 15 años y la tasa de natalidad
correlacion_poblacion_menores15_natalidad, _ = pearsonr(poblacion_menores15, tasa_natalidad)
print("Correlación entre población menor de 15 años y tasa de natalidad normalizada:", correlacion_poblacion_menores15_natalidad)
```

```python
# Definir el diccionario para almacenar los datos
datos_con_porcentaje = {}

# Calcular y agregar el porcentaje de población menor de 15 años sobre la población total para cada año
for año, valores in datos_combinados.items():
    poblacion_menores_15 = valores['Población Menores de 15']
    poblacion_total = valores['Población Total']
    porcentaje_menores_15 = (poblacion_menores_15 / poblacion_total) * 100
    valores['Porcentaje Población Menor de 15'] = porcentaje_menores_15
    datos_con_porcentaje[año] = valores

# Mostrar los resultados con el nuevo valor agregado
for año, valores in datos_con_porcentaje.items():
    print(año)
    print(valores)
    print()
```
